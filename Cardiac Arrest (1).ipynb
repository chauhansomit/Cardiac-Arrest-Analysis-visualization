```python
#  Function to classify ages into age groups

def age_group(age):
  if age < 30:
    return 'Below 30'
  elif age <=50:
    return "30 to 50"
  else:
    return "Above 50"
```

```python
# Apply the age_group function to the 'Age' column and create a new 'age_group' column

data['age_group'] = data['Age'].apply(age_group)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(
    x = 'age_group',
    y = 'Cardiac_Arrest',
    data = age,
    palette = 'muted'

    )
```

```python
def Sleep(Sleep):
  if Sleep < 3:
    return 'Below 2'
  elif Sleep <=5:
    return "3 to 5"
  elif Sleep <= 8:
    return "5 to 8"
  elif Sleep <= 10:
    return '8 to 10'
  else:
    return "Above 10"
```

```python
data['Sleep_Group'] = data['Sleep_Hours'].apply(Sleep)
```

```python
data['Sleep_Group'].unique()
```

# Data Analysis, Data Visualization

```python
# Group data by 'Gender' and calculate the total of 'Family_History' and 'Cardiac_Arrest' for each gender

gender = data.groupby('Gender')[['Family_History','Cardiac_Arrest']].sum().reset_index()
```

```python
# View aggregated data by Gender

gender
```

```python
plt.figure(figsize=(6,4))
sns.barplot(
    x = 'Gender',
    y = 'Family_History',
    data = gender,
    palette = 'muted'

    )

plt.title('Family History by Gender')
plt.xlabel('Gender')
plt.ylabel('Family History')
plt.show()


plt.figure(figsize=(6,4))
sns.barplot(
    x = 'Gender',
    y = 'Cardiac_Arrest',
    data = gender,
    palette = 'muted'
  )

plt.title('Cardiac Arrest by Gender')

plt.show()
```

```python
# Sleep Group-wise Record Count

Sleep = data.groupby('Sleep_Group')['Sleep_Group'].count()
```

```python
Sleep
```

```python
plt.figure(figsize=(7,5))
sns.countplot(
    x = 'Sleep_Group',
    data = data,
    palette ='Set1'
)
plt.title('Count of Individuals by Sleep Group')

plt.xlabel('Sleep Tiem Group')
```

```python
# Alcohol and Smoking Usage Based on Physical Activity Level

drugs_uses = data.groupby('Physical_Activity_Level')[['Alcohol_Intake','Smoking']].sum().reset_index()
```

```python
drugs_uses
```

```python


plt.figure(figsize=(6,4))

plt.plot(
    drugs_uses['Physical_Activity_Level'],
    drugs_uses['Alcohol_Intake'],
    marker = 'o',
    c = 'red',
    linestyle = '--'

    )

plt.plot(
    drugs_uses['Physical_Activity_Level'],
    drugs_uses['Smoking'],
    marker = '*',
    c = 'blue',

)

plt.title('Alcohol and Smoking Usage Based on Physical Activity Level')

# Smoking = Blue
# Alcohol_Intake = Red
```
